# Create Toy Dataset

```python
import rasterio
import pandas as pd
import numpy as np
```

```python
nBands = 8
nRows, nCols = 100, 100
nSamples = 32
pathDir = '../share/data/'
```

```python
datafile = []

# Train samples
for i in range(nSamples):
    data = np.random.rand(nRows, nCols, nBands)
    path = f'{pathDir}/train/{i:02d}.tif'

    with rasterio.open(path, 'w', driver='GTiff', height=nRows, width=nCols, count=nBands, dtype=data.dtype) as dst:
        for b in range(nBands):
            dst.write(data[:, :, b], b + 1)

    datafile.append({
        'embeddings': path,
        'ratio': np.random.uniform(),
        'train': 1
    })

# Test samples
for i in range(nSamples // 4):
    data = np.random.rand(nRows, nCols, nBands)
    path = f'{pathDir}/test/{i:02d}.tif'

    with rasterio.open(path, 'w', driver='GTiff', height=nRows, width=nCols, count=nBands, dtype=data.dtype) as dst:
        for b in range(nBands):
            dst.write(data[:, :, b], b + 1)
    
    datafile.append({
        'embeddings': path,
        'ratio': None,
        'train': 0
    })

datafile = pd.DataFrame(datafile)
datafile.to_csv(f'{pathDir}/data.csv', index=False)
datafile.head()
```
