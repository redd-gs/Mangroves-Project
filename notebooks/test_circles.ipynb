# Test Circles

```python
import math
import numpy as np
import matplotlib.pyplot as plt
from mangroves.utils import haversine, geodesic_circle, planar_approx_circle
```

```python
lat0_deg = 1.41150
lon0_deg = 103.9569

lat1_deg = lat0_deg + 0.0005
lon1_deg = lon0_deg + 0.0006

distance = haversine(lat0_deg, lat1_deg, lon0_deg, lon1_deg)
print(f"Haversine distance: {distance:.2f} meters")

circle_geodesic = geodesic_circle(lat0_deg, lon0_deg, distance)
circle_geodesic_lats, circle_geodesic_lons = zip(*circle_geodesic)

plt.figure(figsize=(3, 3))
plt.plot(circle_geodesic_lons, circle_geodesic_lats, label='Geodesic Circle', color='blue')
plt.scatter([lon0_deg, lon1_deg], [lat0_deg, lat1_deg], color='red', label='Points')
plt.text(lon0_deg, lat0_deg, 'Point 1', fontsize=9, ha='right')
plt.text(lon1_deg, lat1_deg, 'Point 2', fontsize=9, ha='right')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geodesic Circle around Point 1')
plt.axis('equal')
plt.legend(loc ='upper left')
plt.grid()
```

```python
lat0_deg = 1.41150
lon0_deg = 103.9569

lat1_deg = lat0_deg + 0.0005
lon1_deg = lon0_deg + 0.0006

distance = haversine(lat0_deg, lat1_deg, lon0_deg, lon1_deg)
print(f"Haversine distance: {distance:.2f} meters")

circle_geodesic = geodesic_circle(lat0_deg, lon0_deg, distance)
circle_geodesic_lats, circle_geodesic_lons = zip(*circle_geodesic)

circle_approx = planar_approx_circle(lat0_deg, lon0_deg, distance)
circle_approx_lats, circle_approx_lons = zip(*circle_approx)

plt.figure(figsize=(3, 3))
plt.plot(circle_geodesic_lons, circle_geodesic_lats, label='Geodesic Circle', color='blue')
plt.plot(circle_approx_lons, circle_approx_lats, label='Planar Approx Circle', color='green', linestyle='--')
plt.scatter([lon0_deg, lon1_deg], [lat0_deg, lat1_deg], color='red', label='Points')
plt.text(lon0_deg, lat0_deg, 'Point 1', fontsize=9, ha='right')
plt.text(lon1_deg, lat1_deg, 'Point 2', fontsize=9, ha='right')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Geodesic Circle around Point 1')
plt.axis('equal')
plt.legend(loc ='upper left')
plt.grid()
```
